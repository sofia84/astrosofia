```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dati_excel= 'dati.txt'
df_excel=pd.read_csv(dati_excel,
                     sep = '\t',
                     header=None,
                     decimal=',',
                     skiprows=1)
```

```python
df_excel.rename(columns={0:"delta elevation", 1: 'rad'}, inplace = True)
df_excel = df_excel.iloc[:,[0, 1]]
df_excel = df_excel.at[]
```
